```python
import pandas as pd
import numpy as np
from datetime import timedelta, date
```

```python
server = 'server_0'
```

```python
problem_list = ['problem_' + str(num) for num in range(3)]
problem_list
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
start_date = date(2020, 1, 1)
end_date = date(2020, 2, 1)
```

```python
def error_generator(start_date, end_date):
    # Set seed for reproducibility
    np.random.seed(102)
    
    df = pd.DataFrame(columns=['server', 'problem_0', 'problem_1', 'problem_2', 'date'])
    daterange_list = daterange(start_date, end_date)
    for day in daterange_list:
        day_df = pd.DataFrame(columns=['server', 'problem_0', 'problem_1', 'problem_2', 'date'])
        day_df['server'] = [server]
        day_df['date'] = [day]
        for problem in problem_list:
            if problem == 'problem_2':
                problem_count = int(abs(np.round(np.random.normal(scale=10))))
                day_df[problem] = [problem_count]
                continue
            if abs(np.random.normal()) > 0.9:
                day_df[problem] = [1]
            else:
                day_df[problem] = [0]
        df = df.append(day_df, ignore_index=True, sort=True)
    return df.reset_index(drop=True)
```

```python
error_data = error_generator(start_date, end_date)
```
